```python
class Order:
    """
    Order 类：
      - order_id: 订单唯一编号
      - creation_time: 订单创建时间
      - sender_id, receiver_id: 发货/收货方 ID
      - material_id: 物料类型 ("product" 或 "raw_material")
      - quantity: 需求量
      - order_type: "product" 或 "raw_material"
      - distance: 发货方与收货方的距离
      - shipped: 是否已从发货方仓库出库
      - completed: 是否已最终送达收货方
      - theoretical_delivery_time: 理论送达时间 = distance / LOGISTICS_SPEED
      - remaining_time: 初始即设为 theoretical_delivery_time；只有在订单发货后才开始倒计时
      - actual_delivery_time: 订单完成时，实际耗时 = 当前时间步 - creation_time
    """
    def __init__(self, order_id, creation_time, sender_id, receiver_id,
                 material_id, quantity, order_type, distance):
        self.order_id = order_id
        self.creation_time = creation_time
        self.sender_id = sender_id
        self.receiver_id = receiver_id
        self.material_id = material_id
        self.quantity = quantity
        self.order_type = order_type
        self.distance = distance

        self.shipped = False     # 是否已出库
        self.completed = False   # 是否已送达

        # 理论送达时间，单位：时间步
        self.theoretical_delivery_time = distance / LOGISTICS_SPEED
        # 初始 remaining_time 就等于理论值
        self.remaining_time = self.theoretical_delivery_time

        # actual_delivery_time 从订单创建时开始计时
        self.actual_delivery_time = None

    def start_shipping(self, current_step):
        """
        当订单真正发货时调用：
          - 标记 shipped 为 True
          - 此时 remaining_time 开始倒计时
        """
        self.shipped = True
    def update(self, current_step):
        """
        每个时间步由模型调用：
          - 若订单已发货但未完成，则 remaining_time 递减 1；
          - 当 remaining_time <= 0 时，标记订单完成，
            并计算 actual_delivery_time = current_step - creation_time。
          - 若订单尚未发货，则 remaining_time 保持不变。
        """
        if self.shipped and not self.completed:
            self.remaining_time -= 1
            if self.remaining_time <= 0:
                self.remaining_time = 0
                self.completed = True
                self.actual_delivery_time = current_step - self.creation_time

    def __repr__(self):
        return (f"Order({self.order_id}, type={self.order_type},quantity={self.quantity},shipped={self.shipped}, "
                f"completed={self.completed}, remaining_time={self.remaining_time:.1f}, "
                f"theoretical={self.theoretical_delivery_time:.1f})")
```

```python
class Customer(Agent):
    """
    Customer 代理：
      - 每个时间步以一定概率产生一个产品订单（"product"），
      - 随机选择状态为 'Y' 的 Manufacturer 下单，
      - 将订单加入模型的 pending_deliveries 供后续 update()。
    """
    def __init__(self, unique_id, model, pos, demands_list, demand_probability=0.4):
        super().__init__(unique_id, model)
        self.pos = pos
        self.demands_list = demands_list
        self.demand_probability = demand_probability
        self.order_counter = 0
        self.orders = {}  # 存储该 customer 发出的订单

    def generate_order(self):
        """以一定概率产生订单，创建 Order 对象并发送给 Manufacturer"""
        if random.random() < self.demand_probability:
            demand = random.choice(self.demands_list)
            self.order_counter += 1
            
            # 从模型中筛选状态为 'Y' 的 Manufacturer
            valid_manufacturers = [
                agent for agent in self.model.schedule.agents
                if isinstance(agent, Manufacturer) and agent.status == 'Y'
            ]
            if valid_manufacturers:
                chosen_manufacturer = random.choice(valid_manufacturers)
                order_id = f"{self.unique_id}_{self.order_counter}_{chosen_manufacturer.unique_id}"
                # 计算两者之间的距离
                distance = manhattan_distance(self.pos, chosen_manufacturer.pos)
                
                # 创建订单
                order = Order(
                    order_id=order_id,
                    creation_time=self.model.schedule.steps,
                    sender_id=self.unique_id,
                    receiver_id=chosen_manufacturer.unique_id,
                    material_id="product",
                    quantity=demand,
                    order_type="product",
                    distance=distance
                )

                self.orders[order_id] = order
                chosen_manufacturer.receive_order(order)
                # 将订单加入全局列表
                self.model.pending_deliveries.append(order)
            else:
                # 没有可用的 Manufacturer
                pass

    def step(self):
        self.generate_order()

    def receive_delivery(self, order):
        """当订单最终送达时，可在此处记录或打印日志"""
        print(f"Customer {self.unique_id} received {order.order_id} at step {self.model.schedule.steps}, "
              f"actual_delivery_time={order.actual_delivery_time}, "
              f"theoretical={order.theoretical_delivery_time:.1f}")
```

```python
class Manufacturer(Agent):
    """
      1. 先来先发 (FIFO) 的发货逻辑，不进行部分交付。
      2. 排产逻辑：若 gap < 0，需要生产。1 产品需 2 单位原材料。- 若原材料不足，则消耗全部原材料进行部分生产，并向供应商下订单。
      3. 连续 5 个时间步 gap < 0，则 status = 'N'；否则 'Y'。
      4. 库存在发货和生产时实时更新：finished_goods_inventory, raw_material_inventory。
      5. 加入库存上限 (inventory_capacity)：每个时间步成品库存不能超过该上限，
         生产时也只能生产到库存达到上限为止。
    """
    def __init__(self, unique_id, model, pos, production_capacity=100,
                 initial_product_inventory=50, initial_raw_material_inventory=50, inventory_capacity=500):
        super().__init__(unique_id, model)
        self.pos = pos
        self.production_capacity = production_capacity
        self.status = 'Y'
        
        self.finished_goods_inventory = initial_product_inventory
        self.raw_material_inventory = initial_raw_material_inventory
        # 库存上限（成品库存上限）
        self.inventory_capacity = inventory_capacity
        
        self.pending_orders = {}
        self.gap_history = []
        
        self.current_produced = 0
        self.current_shipped = 0

         # 新增成本累积属性
        self.cumulative_storage_cost = 0      # 累计仓储成本，每步成本 = finished_goods_inventory * 1
        self.cumulative_transport_cost = 0    # 累计运输成本，订单完成时计算
        self.cumulative_penalty_cost = 0      # 累计违约成本，订单完成时计算
        self.completed_products = 0           # 累计完成发货的产品数量

    def receive_order(self, order):
        if order.order_type == "product":
            self.pending_orders[order.order_id] = order

    def ship_orders(self):
        # 先来先发
        sorted_orders = sorted(self.pending_orders.values(), key=lambda o: o.creation_time)
        shipped_ids = []
        for order in sorted_orders:
            if not order.shipped:
                if self.finished_goods_inventory >= order.quantity:
                    # 扣减库存
                    self.finished_goods_inventory -= order.quantity
                    self.current_shipped += order.quantity
                    # 调用订单的 start_shipping，而非直接 order.shipped=True
                    order.start_shipping(self.model.schedule.steps)
                    shipped_ids.append(order.order_id)
                else:
                    break
        
        for s_id in shipped_ids:
            del self.pending_orders[s_id]

    def calculate_gap(self):
        backlog_demand = sum(o.quantity for o in self.pending_orders.values() if not o.shipped)
        gap = self.finished_goods_inventory - backlog_demand
        return gap

    def update_status(self, gap):
        self.gap_history.append(gap)
        if len(self.gap_history) > 20:
            self.gap_history.pop(0)
        if len(self.gap_history) == 20 and all(g < 0 for g in self.gap_history):
            self.status = 'N'
        else:
            self.status = 'Y'
            
    def update_incoming_raw_material_orders(self):
        """
        检查全局 pending_deliveries 中所有属于本制造商下的原材料订单，
        若订单完成（completed==True），则将该订单的数量加到 raw_material_inventory 中，
        并从 pending_deliveries 中移除，记录到 delivered_orders 中。
        """
        for order in list(self.model.pending_deliveries):
            # 判断该订单是否为原材料订单，且是由本制造商下的订单
            if order.order_type == "raw_material" and order.sender_id == self.unique_id:
                # 更新订单状态（倒计时）
                order.update(self.model.schedule.steps)
                if order.completed:
                    self.raw_material_inventory += order.quantity
                    self.model.delivered_orders.append(order)
                    self.model.pending_deliveries.remove(order)

    def production_logic(self):
        """
        排产逻辑：
          1. 计算 gap = finished_goods_inventory - backlog_demand
             若 gap < 0，则需要生产数量 production_needed = min(|gap|, production_capacity)
          2. 同时考虑库存上限：可生产量不能超过 available_space = inventory_capacity - finished_goods_inventory
          3. 1 产品需 2 单位原材料：
             - 若原材料充足，则一次性生产 production_qty；
             - 若原材料不足，则消耗全部原材料进行部分生产，剩余部分通过下单补充。
        """
        gap = self.calculate_gap()
        if gap < 0:
            # 1 产品需要 2 单位原材料，计算为满足产品缺口需要的原材料数量
            required_rm_for_gap = 2 * abs(gap)       
            # 计算在途的原材料订单数量（只考虑本制造商下的原材料订单）
            in_transit_rm = sum(o.quantity for o in self.model.pending_deliveries 
                                if o.order_type == "raw_material" and o.sender_id == self.unique_id)
            net_rm = self.raw_material_inventory + in_transit_rm
            
            # 如果净原材料供应不足，则下采购单补齐差额（不受生产能力、库存上限限制）,+30%采购buffer
            if net_rm < 1.8* required_rm_for_gap:
                shortage = 1.8* required_rm_for_gap - net_rm
                self.order_raw_material(shortage)

            
            # 理论上需要生产的产品数量
            production_needed = min(1.6* abs(gap), self.production_capacity)
            # 考虑库存上限：本步最多能生产的数量
            available_space = self.inventory_capacity - self.finished_goods_inventory
            # 实际生产数量受两者限制
            production_needed = min(production_needed, available_space)
            required_rm = 2 * production_needed
            if self.raw_material_inventory >= required_rm:
                # 原材料充足，一次性生产
                self.finished_goods_inventory += production_needed
                self.raw_material_inventory -= required_rm
                self.current_produced = production_needed
            else:
                # 原材料不足，消耗全部原材料进行部分生产
                partial_production = self.raw_material_inventory // 2
                self.finished_goods_inventory += partial_production
                self.raw_material_inventory -= partial_production * 2
                self.current_produced = partial_production
                short_production = production_needed - partial_production
        else:
            self.current_produced = 0

    def order_raw_material(self, raw_material_needed):
        valid_suppliers = [
            agent for agent in self.model.schedule.agents
            if isinstance(agent, Supplier) and agent.status == 'Y'
        ]
        if valid_suppliers:
            chosen_supplier = random.choice(valid_suppliers)
            order_id = f"{self.unique_id}_RM_{self.model.schedule.steps}"
            dist = manhattan_distance(self.pos, chosen_supplier.pos)
            raw_order = Order(
                order_id=order_id,
                creation_time=self.model.schedule.steps,
                sender_id=self.unique_id,
                receiver_id=chosen_supplier.unique_id,
                material_id="raw_material",
                quantity=raw_material_needed,
                order_type="raw_material",
                distance=dist
            )
            chosen_supplier.receive_order(raw_order)
            self.model.pending_deliveries.append(raw_order)

    def step(self):
        self.update_incoming_raw_material_orders()
        self.ship_orders()
        gap = self.calculate_gap()
        self.update_status(gap)
        self.production_logic()
        self.current_shipped = 0 
        # 累计当前时刻的仓储成本（单位仓储成本为1）
        self.cumulative_storage_cost += self.finished_goods_inventory*1
```

```python
def update_transport_cost_for_order(order, model):
    if order.order_type == "product" and order.completed:
        # 计算运输成本：实际送达天数 * 订单数量
        transport_cost = order.actual_delivery_time * order.quantity
        # 计算违约成本：若实际送达天数大于理论送达天数，则额外惩罚 = 延误天数 * 订单数量 * 1
        penalty_cost = 0
        if order.actual_delivery_time > order.theoretical_delivery_time:
            delay = order.actual_delivery_time - order.theoretical_delivery_time
            penalty_cost = delay * order.quantity * 5
        # 查找对应的 Manufacturer（订单的 receiver_id）
        for m in model.manufacturers:
            if m.unique_id == order.receiver_id:
                m.cumulative_transport_cost += transport_cost
                m.cumulative_penalty_cost += penalty_cost
                m.completed_products += order.quantity
                break
```

```python
class Supplier(Agent):
    """
    Supplier 代理，逻辑与 Manufacturer 一致：
      1. 存储未发货订单 pending_orders，先来先发 (FIFO) 不部分交付。
      2. 使用 raw_material_inventory 表示可生产或已产出的原材料库存。
      3. 若 gap < 0 则生产 (不需向上游下单)，生产量 = min(abs(gap), production_capacity)。
      4. 连续 5 个时间步 gap < 0，则 status='N'，否则='Y'。
    """
    def __init__(self, unique_id, model, pos,
                 material_capacity=100,
                 initial_raw_material_inventory=50):
        super().__init__(unique_id, model)
        self.pos = pos
        self.material_capacity = material_capacity
        self.status = 'Y'
        
        self.raw_material_inventory = initial_raw_material_inventory
        self.pending_orders = {}
        self.gap_history = []
        
        self.current_produced = 0
        self.current_shipped = 0

    def receive_order(self, order):
        if order.order_type == "raw_material":
            self.pending_orders[order.order_id] = order

    def ship_orders(self):
        sorted_orders = sorted(self.pending_orders.values(), key=lambda o: o.creation_time)
        shipped_ids = []
        for order in sorted_orders:
            if not order.shipped:
                if self.raw_material_inventory >= order.quantity:
                    self.raw_material_inventory -= order.quantity
                    self.current_shipped += order.quantity
                    # 开始发货
                    order.start_shipping(self.model.schedule.steps)
                    shipped_ids.append(order.order_id)
                else:
                    break
        for s_id in shipped_ids:
            del self.pending_orders[s_id]

    def calculate_gap(self):
        backlog_demand = sum(o.quantity for o in self.pending_orders.values() if not o.shipped)
        return self.raw_material_inventory - backlog_demand

    def update_status(self, gap):
        self.gap_history.append(gap)
        if len(self.gap_history) > 10:
            self.gap_history.pop(0)
        if len(self.gap_history) == 10 and all(g < 0 for g in self.gap_history):
            self.status = 'N'
        else:
            self.status = 'Y'

    def production_logic(self):
        gap = self.calculate_gap()
        if gap < 0:
            production_needed = min(abs(gap), self.material_capacity)
            self.raw_material_inventory += production_needed
            self.current_produced = production_needed
        else:
            self.current_produced = 0

    def step(self):
        self.ship_orders()
        gap = self.calculate_gap()
        self.update_status(gap)
        self.production_logic()
        self.current_shipped = 0
```

```python
GRID_SIZE = 10

used_positions = set()

def random_position_unique(pos_list):
    """
    从 pos_list 中随机取一个位置 (x, y)，并将其移除，避免重复使用。
    """
    pos = random.choice(pos_list)
    pos_list.remove(pos)
    return pos

class SupplyChainGridModel(Model):
    def __init__(self,
                 positions,             # (x, y) 坐标列表
                 demands_list,          # 客户需求量列表
                 inventory_list,        # 库存列表
                 num_customers=10,
                 num_manufacturers=4,
                 num_suppliers=2,
                 cust_demand_probability=0.4,
                 m_production_capacity=10,
                 m_inventory_capacity=1000,
                 s_material_capacity=40):
        super().__init__()
        self.schedule = BaseScheduler(self)
        self.pending_deliveries = []
        self.delivered_orders = []
        self.steps = 0

        # 为了后续多次抽取位置不冲突，这里复制一份 positions
        self.available_positions = positions.copy()

        # =========== 创建 Customer ===========
        self.customers = []
        for i in range(num_customers):
            # 从 self.available_positions 随机取一个位置
            c_pos = random_position_unique(self.available_positions)
            # 把 demands_list 直接传给 Customer，后续在 step() 里随机选
            customer = Customer(
                unique_id=f"C{i+1}",
                model=self,
                pos=c_pos,
                demands_list=demands_list,
                demand_probability=cust_demand_probability
            )
            self.customers.append(customer)
            self.schedule.add(customer)

        # =========== 创建 Manufacturer ===========
        self.manufacturers = []
        for i in range(num_manufacturers):
            m_pos = random_position_unique(self.available_positions)
            # 随机从 inventory_list 中取一个值作为初始产品库存
            init_prod_inv = random.choice(inventory_list)
            # 再随机一个值作为初始原材料库存（可根据需求再加工）
            init_raw_mat_inv = random.choice(inventory_list)
            manufacturer = Manufacturer(
                unique_id=f"M{i+1}",
                model=self,
                pos=m_pos,
                production_capacity=m_production_capacity,
                initial_product_inventory=init_prod_inv,
                initial_raw_material_inventory=init_raw_mat_inv,
                inventory_capacity=m_inventory_capacity
            )
            self.manufacturers.append(manufacturer)
            self.schedule.add(manufacturer)

        # =========== 创建 Supplier ===========
        self.suppliers = []
        for i in range(num_suppliers):
            s_pos = random_position_unique(self.available_positions)
            # 随机一个值乘以 3 作为初始原材料库存
            init_raw_material = random.choice(inventory_list) * 3
            supplier = Supplier(
                unique_id=f"S{i+1}",
                model=self,
                pos=s_pos,
                material_capacity=s_material_capacity,
                initial_raw_material_inventory=init_raw_material
            )
            self.suppliers.append(supplier)
            self.schedule.add(supplier)

    def step(self):
        self.schedule.step()
        # 更新所有运输中的订单
        for order in list(self.pending_deliveries):
            if order.order_type == "product":
                order.update(self.schedule.steps)
                if order.completed:
                    receiver_agent = next((a for a in self.schedule.agents if a.unique_id == order.receiver_id), None)
                    if receiver_agent:
                        if isinstance(receiver_agent, Customer):
                            receiver_agent.receive_delivery(order)

                    # **这里**更新制造商的统计信息
                    update_transport_cost_for_order(order, self)
                    self.delivered_orders.append(order)
                    self.pending_deliveries.remove(order)
        self.steps += 1

# 示例运行
if __name__ == "__main__":
    model = SupplyChainGridModel(positions=positions,
        demands_list=demands_list,
        inventory_list=inventory_list,
        num_customers=10,
        num_manufacturers=3,
        num_suppliers=2,
        cust_demand_probability=0.3,
        m_production_capacity=10,
        m_inventory_capacity=200,
        s_material_capacity=40)
    total_steps = 1000
    # 用于记录每50步的统计数据
    stats = []
    for t in range(1, total_steps+1):
        model.step()
        if t % 50 == 0:
            # 分别统计 product 订单和 raw_material 订单
            prod_orders = [o for o in model.delivered_orders if o.order_type == "product"]
            raw_orders  = [o for o in model.delivered_orders if o.order_type == "raw_material"]
            sum_actual_prod = sum(o.actual_delivery_time for o in prod_orders if o.actual_delivery_time is not None)
            sum_theoretical_prod = sum(o.theoretical_delivery_time for o in prod_orders)
            ratio_prod = sum_actual_prod / sum_theoretical_prod if sum_theoretical_prod > 0 else None
        
            sum_actual_raw = sum(o.actual_delivery_time for o in raw_orders if o.actual_delivery_time is not None)
            sum_theoretical_raw = sum(o.theoretical_delivery_time for o in raw_orders)
            ratio_raw = sum_actual_raw / sum_theoretical_raw if sum_theoretical_raw > 0 else None
        
            stats.append((t, ratio_prod, ratio_raw))
            prod_str = f"{ratio_prod:.3f}" if ratio_prod is not None else "N/A"
            raw_str = f"{ratio_raw:.3f}" if ratio_raw is not None else "N/A"
            print(f"Step {t}: Product ratio = {prod_str} ; Raw material ratio = {raw_str}")
            
            all_orders = model.pending_deliveries + model.delivered_orders
            summary = {}
            for order in all_orders:
                typ = order.order_type
                if typ not in summary:
                    summary[typ] = {"not_shipped": 0, "in_transit": 0, "completed": 0}
                if not order.shipped:
                    summary[typ]["not_shipped"] += 1
                elif order.shipped and not order.completed:
                    summary[typ]["in_transit"] += 1
                elif order.completed:
                    summary[typ]["completed"] += 1
            # 输出统计信息
            for order_type, counts in summary.items():
                print(f"Order type: {order_type} -> Not shipped: {counts['not_shipped']}, "
                      f"In transit: {counts['in_transit']}, Completed: {counts['completed']}")
```

```python
LOGISTICS_SPEED = 2
class SupplyChainGridModel(Model):
    def __init__(self, positions,             
                 demands_list,          
                 inventory_list,        
                 num_customers=10,
                 num_manufacturers=6,
                 num_suppliers=2,
                 cust_demand_probability=0.3,
                 m_production_capacity=m_prod,
                 m_inventory_capacity=1000,
                 s_material_capacity=40, heterogeneous=False):
        super().__init__()
        self.schedule = BaseScheduler(self)
        self.pending_deliveries = []
        self.delivered_orders = []
        self.steps = 0
        self.heterogeneous = heterogeneous
        self.available_positions = positions.copy()
        self.customers = []
        for i in range(num_customers):
            customer = Customer(f"C{i+1}", self, pos=random_position_unique(self.available_positions),
                    demands_list=demands_list, demand_probability=cust_demand_probability)
            self.customers.append(customer)
            self.schedule.add(customer)

        self.manufacturers = []
        for i in range(num_manufacturers):
            if heterogeneous:
                prod_cap = random.choice([max(1, m_production_capacity-5), m_production_capacity+5])
                # 为确保均值接近 m_production_capacity，我们直接随机取[m_production_capacity-5, m_production_capacity+5]
                prod_cap = random.randint(max(1, m_production_capacity-5), m_production_capacity+5)
            else:
                prod_cap = m_production_capacity
            manufacturer = Manufacturer(f"M{i+1}", self, pos=random_position_unique(self.available_positions),
                              production_capacity=prod_cap,
                              initial_product_inventory=random.choice(inventory_list),  # 随机选取产品库存
        initial_raw_material_inventory=random.choice(inventory_list)*2,  # 随机选取原材料库存
                              inventory_capacity=m_inventory_capacity)
            self.manufacturers.append(manufacturer)
            self.schedule.add(manufacturer)

        self.suppliers = []
        for i in range(num_suppliers):
            supplier = Supplier(f"S{i+1}", self, pos=random_position_unique(self.available_positions),
                    material_capacity=s_material_capacity,
                    initial_raw_material_inventory=random.choice(inventory_list)*2)
            self.suppliers.append(supplier)
            self.schedule.add(supplier)

    def step(self):
        self.schedule.step()
        # 模型统一只更新产品订单
        for order in list(self.pending_deliveries):
            if order.order_type == "product":
                order.update(self.schedule.steps)
                if order.completed:
                    receiver_agent = next((a for a in self.schedule.agents if a.unique_id == order.receiver_id), None)
                    if receiver_agent and isinstance(receiver_agent, Customer):
                        receiver_agent.receive_delivery(order)
                    self.delivered_orders.append(order)
                    self.pending_deliveries.remove(order)
        self.steps += 1

# ----------------- 模拟与绘图 -----------------------

def run_simulation(heterogeneous=False, m_prod_value=None, total_steps=1000, num_reps=5):
    """
    若 heterogeneous==False，则采用 uniform 模式，m_prod_value为统一生产能力；
    若 heterogeneous==True，则每个manufacturer随机取值于 [m_prod_value-5, m_prod_value+5]。
    返回：均值、产品订单ratio、原材料订单ratio、制造商状态N的累计频次。
    """
    ratio_prod_list = []
    ratio_mat_list = []
    statusN_freq_list = []
    capacity_means = []
    for rep in range(num_reps):
        used_positions.clear()
        if heterogeneous:
            model = SupplyChainGridModel(positions,             
                 demands_list,          
                 inventory_list,        
                 num_customers=10,
                 num_manufacturers=6,
                 num_suppliers=2,
                 cust_demand_probability=0.4,
                 m_production_capacity=m_prod,
                 m_inventory_capacity=1000,
                 s_material_capacity=300, heterogeneous=True)
            caps = [m.production_capacity for m in model.manufacturers]
            capacity_means.append(np.mean(caps))
        else:
            model = SupplyChainGridModel(positions,             
                 demands_list,          
                 inventory_list,        
                 num_customers=10,
                 num_manufacturers=6,
                 num_suppliers=2,
                 cust_demand_probability=0.4,
                 m_production_capacity=m_prod,
                 m_inventory_capacity=1000,
                 s_material_capacity=300, heterogeneous=False)
            capacity_means.append(m_prod_value)
        statusN_freq = 0
        for t in range(1, total_steps+1):
            model.step()
            statusN_freq += sum(1 for m in model.manufacturers if m.status == 'N')
        prod_orders = [o for o in model.delivered_orders if o.order_type == "product" and o.actual_delivery_time is not None]
        sum_actual_prod = sum(o.actual_delivery_time for o in prod_orders)
        sum_theoretical_prod = sum(o.theoretical_delivery_time for o in prod_orders)
        ratio_prod = sum_actual_prod / sum_theoretical_prod if sum_theoretical_prod > 0 else 0
        ratio_prod_list.append(ratio_prod)
        statusN_freq_list.append(statusN_freq)
    return np.mean(capacity_means), np.mean(ratio_prod_list), np.mean(statusN_freq_list)

# 对于 uniform 模式，m_prod_value 取 5~30 的每个整数
uniform_targets = list(range(5, 30))
uni_caps = []
uni_prod_ratios = []
uni_mat_ratios = []
uni_statusN = []

for m_prod in uniform_targets:
    cap_mean, rprod, statusN = run_simulation(heterogeneous=False, m_prod_value=m_prod, total_steps=1000, num_reps=10)
    uni_caps.append(cap_mean)
    uni_prod_ratios.append(rprod)
    uni_statusN.append(statusN)
    print(f"Uniform m_prod = {m_prod}: cap_mean={cap_mean}, prod ratio={rprod}, status N freq={statusN}")

# 对于 heterogeneous 模式，目标均值取 5~16，每个 manufacturer 随机取值于 [target-5, target+5]
hetero_caps = []
hetero_prod_ratios = []
hetero_statusN = []

for target in uniform_targets:
    reps = 10
    hetero_prod_list = []
    hetero_statusN_list = []
    caps_list = []
    for rep in range(reps):
        used_positions.clear()
        # 对每个 manufacturer，随机取值于 [target-5, target+5]
        num_manufacturers = 6
        manufacturers_caps = [random.randint(max(1, target-10), target+10) for _ in range(num_manufacturers)]
        model = SupplyChainGridModel(positions,             
                 demands_list,          
                 inventory_list,        
                 num_customers=10,
                 num_manufacturers=6,
                 num_suppliers=2,
                 cust_demand_probability=0.4,
                 m_production_capacity=target,
                 m_inventory_capacity=1000,
                 s_material_capacity=300, heterogeneous=False)
        # 替换每个 Manufacturer 的生产能力为随机值
        for i, m in enumerate(model.manufacturers):
            m.production_capacity = manufacturers_caps[i]
        caps_list.append(np.mean(manufacturers_caps))
        statusN_freq = 0
        for t in range(1, 1000+1):
            model.step()
            statusN_freq += sum(1 for m in model.manufacturers if m.status == 'N')
        prod_orders = [o for o in model.delivered_orders if o.order_type == "product" and o.actual_delivery_time is not None]
        sum_actual_prod = sum(o.actual_delivery_time for o in prod_orders)
        sum_theoretical_prod = sum(o.theoretical_delivery_time for o in prod_orders)
        ratio_prod = sum_actual_prod / sum_theoretical_prod if sum_theoretical_prod > 0 else 0
        hetero_prod_list.append(ratio_prod)
        hetero_statusN_list.append(statusN_freq)
    hetero_caps.append(np.mean(caps_list))
    hetero_prod_ratios.append(np.mean(hetero_prod_list))
    hetero_statusN.append(np.mean(hetero_statusN_list))
    print(f"Hetero target = {target}: cap_mean={np.mean(caps_list):.2f}, prod ratio={np.mean(hetero_prod_list):.3f}, status N freq={np.mean(hetero_statusN_list):.1f}")

# 绘制比值图
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
x = np.array(uniform_targets)
plt.errorbar(x, uni_prod_ratios, fmt='o-', label='Uniform Product Ratio')
plt.errorbar(x, hetero_prod_ratios, fmt='o--', label='Hetero Product Ratio')
plt.xlabel('Mean Production Capacity')
plt.ylabel('Ratio (Actual/Theoretical Delivery Time)')
plt.title('Ratio at Step 1000')
plt.legend()
plt.grid(True)

# 绘制状态 N 频次图
plt.subplot(1,2,2)
plt.plot(x, uni_statusN, 'o-', label='Uniform Mode')
plt.plot(x, hetero_statusN, 's--', label='Hetero Mode')
plt.xlabel('Mean Production Capacity')
plt.ylabel('Accumulated Manufacturer N Frequency (1000 steps)')
plt.title('Manufacturer Status N Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def run_simulation_once(heterogeneous, m_prod_value, total_steps=1000):
    """
    只做一次仿真，返回 ratio_prod
    """
    used_positions.clear()
    if heterogeneous:
        model = SupplyChainGridModel(
            positions, demands_list, inventory_list,
            num_customers=10, num_manufacturers=6, num_suppliers=2,
            cust_demand_probability=0.4,
            m_production_capacity=m_prod_value,
            m_inventory_capacity=1000,
            s_material_capacity=200,
            heterogeneous=True
        )
    else:
        model = SupplyChainGridModel(
            positions, demands_list, inventory_list,
            num_customers=10, num_manufacturers=6, num_suppliers=2,
            cust_demand_probability=0.4,
            m_production_capacity=m_prod_value,
            m_inventory_capacity=1000,
            s_material_capacity=200,
            heterogeneous=False
        )
    
    for t in range(total_steps):
        model.step()
    # 计算产品订单 ratio
    prod_orders = [o for o in model.delivered_orders
                   if o.order_type == "product" and o.actual_delivery_time is not None]
    sum_actual_prod = sum(o.actual_delivery_time for o in prod_orders)
    sum_theoretical_prod = sum(o.theoretical_delivery_time for o in prod_orders)
    ratio_prod = sum_actual_prod / sum_theoretical_prod if sum_theoretical_prod > 0 else 0
    return ratio_prod

def run_experiment(heterogeneous, total_steps=1000, num_reps=10):
    """
    对 m_production_capacity=5..29 依次做多次重复试验（num_reps 次），
    返回每个产能下的 [mean, std]。
    """
    m_values = range(10, 51, 2)
    all_means = []
    all_stds = []
    for m_prod in m_values:
        replicate_ratios = []
        for rep in range(num_reps):
            ratio = run_simulation_once(heterogeneous, m_prod_value=m_prod, total_steps=total_steps)
            replicate_ratios.append(ratio)
        mean_val = np.mean(replicate_ratios)
        std_val = np.std(replicate_ratios)
        all_means.append(mean_val)
        all_stds.append(std_val)
    return list(m_values), all_means, all_stds

# 分别获取 Uniform 和 Heterogeneous 模式的数据
m_values, uni_means, uni_stds = run_experiment(heterogeneous=False, total_steps=1000, num_reps=10)
_, het_means, het_stds = run_experiment(heterogeneous=True, total_steps=1000, num_reps=10)
plt.figure(figsize=(10,6))

x = np.array(m_values)

# 1) Uniform
uni_means = np.array(uni_means)
uni_stds = np.array(uni_stds)
plt.plot(x, uni_means, 'o-', label='Uniform Product Ratio', color='blue')
plt.fill_between(x,
                 uni_means - uni_stds,
                 uni_means + uni_stds,
                 alpha=0.2, color='blue')

# 2) Heterogeneous
het_means = np.array(het_means)
het_stds = np.array(het_stds)
plt.plot(x, het_means, 's--', label='Hetero Product Ratio', color='orange')
plt.fill_between(x,
                 het_means - het_stds,
                 het_means + het_stds,
                 alpha=0.2, color='orange')

plt.xlabel('m_production_capacity')
plt.ylabel('Ratio (Actual/Theoretical Delivery Time)')
plt.title('Comparison: Uniform vs Heterogeneous (with shading)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import networkx as nx

def get_weighted_edge_count(model):
    """
    构建一个 MultiGraph，将模型中所有代理作为节点，
    对于每个 delivered_orders 中已完成且已发货的订单，
    在发货方和收货方之间添加一条边，边的权重为订单数量。
    返回图中所有边的权重和，即为 weighted edge count。
    """
    G = nx.MultiGraph()
    for agent in model.schedule.agents:
        G.add_node(agent.unique_id)
    for order in model.delivered_orders:
        if order.completed and order.shipped and order.order_type == "product":
            # 添加边时指定 weight 属性为订单数量
            G.add_edge(order.sender_id, order.receiver_id, weight=order.quantity)
    total_weight = sum(data.get('weight', 1) for u, v, data in G.edges(data=True))
    return total_weight

# 运行仿真 total_steps 步后返回网络的 weighted edge count
def run_simulation_and_get_weighted_edge_count(heterogeneous, m_prod_value, total_steps=1000):
    if heterogeneous:
        model = SupplyChainGridModel(positions, demands_list, inventory_list,
                                     num_customers=10,
                                     num_manufacturers=6,
                                     num_suppliers=2,
                                     cust_demand_probability=0.3,
                                     m_production_capacity=m_prod_value,
                                     m_inventory_capacity=1000,
                                     s_material_capacity=300,
                                     heterogeneous=True)
    else:
        model = SupplyChainGridModel(positions, demands_list, inventory_list,
                                     num_customers=10,
                                     num_manufacturers=6,
                                     num_suppliers=2,
                                     cust_demand_probability=0.3,
                                     m_production_capacity=m_prod_value,
                                     m_inventory_capacity=1000,
                                     s_material_capacity=300,
                                     heterogeneous=False)
    for t in range(1, total_steps + 1):
        model.step()
    return get_weighted_edge_count(model)

# 针对不同生产能力值（m_prod_value）分别对两种模式进行仿真，记录平均 weighted edge count
uniform_targets = list(range(5, 41, 2))
num_reps = 10
uniform_weighted_edges = []
for m_prod in uniform_targets:
    edge_weights = []
    for rep in range(num_reps):
        used_positions.clear()  # 清空已使用的位置，确保每次随机分配不重叠
        w = run_simulation_and_get_weighted_edge_count(heterogeneous=False, m_prod_value=m_prod, total_steps=1000)
        edge_weights.append(w)
    avg_weight = np.mean(edge_weights)
    uniform_weighted_edges.append(avg_weight)
    print(f"Uniform mode, m_prod = {m_prod}: average weighted edge count = {avg_weight:.2f}")

hetero_weighted_edges = []
for m_prod in uniform_targets:
    edge_weights = []
    for rep in range(num_reps):
        used_positions.clear()
        w = run_simulation_and_get_weighted_edge_count(heterogeneous=True, m_prod_value=m_prod, total_steps=1000)
        edge_weights.append(w)
    avg_weight = np.mean(edge_weights)
    hetero_weighted_edges.append(avg_weight)
    print(f"Heterogeneous mode, m_prod = {m_prod}: average weighted edge count = {avg_weight:.2f}")

# 绘图：横坐标为生产能力（Production Capacity），纵坐标为平均 weighted edge count
plt.figure(figsize=(8,6))
x = np.array(uniform_targets)
plt.plot(x, uniform_weighted_edges, 'o-', label='Uniform Mode')
plt.plot(x, hetero_weighted_edges, 's--', label='Heterogeneous Mode')
plt.xlabel('Production Capacity')
plt.ylabel('Average Weighted Edge Count')
plt.title('Weighted Edge Count vs Production Capacity (1000 steps)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def simulate_weighted_edge_time(heterogeneous, m_prod_value, total_steps=1000):
    """
    在给定的 heterogeneous 模式和生产能力（m_prod_value）下，
    运行 total_steps 步仿真，记录每个时间步的加权边数：
      - completed_weight：已完成且已出库订单的加权和
      - transit_weight：已出库但未完成订单（运输中）的加权和
    返回：时间步列表、completed_weight 列表、transit_weight 列表
    """
    # 创建模型（这里采用 heterogeneous 模式，目标生产能力为 m_prod_value）
    model = SupplyChainGridModel(positions, demands_list, inventory_list,
                                 num_customers=10,
                                 num_manufacturers=6,
                                 num_suppliers=2,
                                 cust_demand_probability=0.4,
                                 m_production_capacity=m_prod_value,
                                 m_inventory_capacity=1000,
                                 s_material_capacity=200,
                                 heterogeneous=heterogeneous)
    
    # 用于记录每一步的加权边数
    time_steps = []
    completed_weights = []  # 来自 delivered_orders
    transit_weights = []    # 来自 pending_deliveries（已出库但未完成）
    
    for t in range(1, total_steps + 1):
        model.step()
        time_steps.append(t)
        # 计算已完成订单的加权和
        comp_w = sum(order.quantity for order in model.delivered_orders if order.completed and order.shipped)
        # 计算运输中订单的加权和
        trans_w = sum(order.quantity for order in model.pending_deliveries if order.shipped and not order.completed)
        completed_weights.append(comp_w)
        transit_weights.append(trans_w)
    return time_steps, completed_weights, transit_weights

# 例如，在 heterogeneous 模式下，目标生产能力取 10
time_steps, comp_w, trans_w = simulate_weighted_edge_time(heterogeneous=True, m_prod_value=10, total_steps=1000)

# 绘图：画出 completed_weight 和 transit_weight 随时间步变化的曲线
plt.figure(figsize=(10,6))
plt.plot(time_steps, comp_w, label='Completed Orders Weighted Edge Count', color='blue')
plt.plot(time_steps, trans_w, label='Transit Orders Weighted Edge Count', color='red')
plt.xlabel('Time Step')
plt.ylabel('Weighted Edge Count (Order Quantity)')
plt.title('Weighted Edge Count vs Time (Heterogeneous Mode, m_prod=10)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def simulate_weighted_edge_time(heterogeneous, m_prod_value, total_steps=1000):
    """
    单次仿真：
      - 每步记录已完成订单的累积加权和 (comp_w)、
        运输中订单的加权和 (trans_w)
      - 返回 time_steps, comp_w, trans_w
    """
    model = SupplyChainGridModel(positions, demands_list, inventory_list,
                                 num_customers=10,
                                 num_manufacturers=6,
                                 num_suppliers=2,
                                 cust_demand_probability=0.4,
                                 m_production_capacity=m_prod_value,
                                 m_inventory_capacity=1000,
                                 s_material_capacity=200,
                                 heterogeneous=heterogeneous)
    
    time_steps = []
    completed_weights = []
    transit_weights = []
    
    for t in range(1, total_steps + 1):
        model.step()
        time_steps.append(t)
        # 已完成订单的加权和（累积）
        comp_w = sum(order.quantity for order in model.delivered_orders 
                     if order.completed and order.shipped)
        # 运输中订单的加权和
        trans_w = sum(order.quantity for order in model.pending_deliveries 
                      if order.shipped and not order.completed)
        
        completed_weights.append(comp_w)
        transit_weights.append(trans_w)
    
    return time_steps, completed_weights, transit_weights

def run_experiment_new_completed(heterogeneous, m_prod_value, total_steps=1000, num_reps=10):
    """
    重复 num_reps 次仿真，每次运行 total_steps 步，
    计算每个时间步“新完成订单加权数” (comp_w[i] - comp_w[i-1])，
    最后对10次结果取平均。
    返回 time_steps, new_completed_mean（长度 total_steps）
    """
    # 用于存储“新完成订单增量”序列
    # replicate_new_comp[r][i] 表示第 r 次试验在第 i 步的新增完成订单加权数
    replicate_new_comp = []
    
    for rep in range(num_reps):
        used_positions.clear()  # 重新分配位置
        time_steps, comp_w, trans_w = simulate_weighted_edge_time(heterogeneous, m_prod_value, total_steps)
        
        # 计算“当步新增完成”的序列
        # comp_w[i]是第i步结束的累积完成加权数
        # new_comp[0]可设为 comp_w[0]，因为没有前一时步
        new_comp = []
        new_comp.append(comp_w[0])  # 第1步新增=comp_w[0]
        for i in range(1, total_steps):
            new_comp.append(comp_w[i] - comp_w[i-1])
        
        replicate_new_comp.append(new_comp)
    
    # 现在对 replicate_new_comp 做逐步平均
    # replicate_new_comp 是 num_reps x total_steps 的二维列表
    new_completed_mean = []
    for i in range(total_steps):
        # 第 i 步的“新完成”在所有rep上的平均
        vals = [replicate_new_comp[r][i] for r in range(num_reps)]
        new_completed_mean.append(np.mean(vals))
    
    # time_steps 与 simulate_weighted_edge_time 对应
    # time_steps 是从 1..total_steps
    # new_completed_mean[i] 对应第 i 步(1-based)
    # return time_steps, new_completed_mean
    return list(range(1, total_steps+1)), new_completed_mean


time_steps, new_comp_mean = run_experiment_new_completed(
    heterogeneous=True, 
    m_prod_value=10, 
    total_steps=1000, 
    num_reps=10
)

# 绘图
plt.figure(figsize=(10,6))
plt.plot(time_steps, new_comp_mean, label='Newly Completed Orders (avg of 10 reps)', color='blue')
plt.xlabel('Time Step')
plt.ylabel('Newly Completed Weighted Count')
plt.title('Newly Completed Weighted Edge Count vs Time\n(Heterogeneous Mode, m_prod=10, 10 reps)')
plt.legend()
plt.grid(True)
plt.show()
```
